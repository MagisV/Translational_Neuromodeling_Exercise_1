# Exercise 5.1 a

1. **Noise distribution**

   $$
   p(\varepsilon)=\sqrt{\frac{\tau}{2\pi}}\,
   \exp\!\left(-\frac{\tau}{2}\varepsilon^{2}\right),
   \qquad
   \log p(\varepsilon)=\tfrac12\log\tau-\tfrac12\log(2\pi)-\tfrac{\tau}{2}\varepsilon^{2}.
   $$

2. **Relate observations to noise**  

   With $y_n=\mu+\varepsilon_n$, each $\varepsilon_n = y_n-\mu$.

3. **Independent observations**

   $$
   \log p(\mathbf y\mid\mu,\tau)
   =\sum_{n=1}^{N}\log p(y_n\mid\mu,\tau)
   =\sum_{n=1}^{N}\left[\frac12\log\tau-\frac12\log(2\pi)-\frac{\tau}{2}(y_n-\mu)^2\right].
   $$

4. **Collect terms**

   $$
   \log p(\mathbf y\mid\mu,\tau)
   =\frac{N}{2}\log\tau
   -\frac{\tau}{2}\sum_{n=1}^{N}(y_n-\mu)^2
   -\frac{N}{2}\log(2\pi)
   $$

$$

$$

From partâ€¯(b) the $\mu$â€‘dependent part of the logâ€‘joint is

Taking the expectation overÂ $\tau$ (i.e.\ replacing $\tau$ by its mean $\mathbb{E}_{\tau}[\tau]$) yields  

$$
%

\,=\,

# 5.1â€¯d

#### 1. Expand the two squared terms  


\begin{aligned}
\sum_{n=1}^{N}(y_n-\mu)^2
&= \sum_{n=1}^{N}\bigl(y_n^2 - 2\mu y_n + \mu^2\bigr)
  = \sum_{n=1}^{N}y_n^2
    -2\mu\sum_{n=1}^{N}y_n
    +N\mu^2, \\[4pt]
\lambda_0(\mu-\mu_0)^2
&= \lambda_0\bigl(\mu^2 -2\mu\mu_0 + \mu_0^2\bigr).
\end{aligned}


#### 2. Collect the $\mu$â€‘dependent pieces  

Group the coefficients of $\mu^2$ and $\mu$:


\begin{aligned}
\sum_{n=1}^{N}(y_n-\mu)^2
  +\lambda_0(\mu-\mu_0)^2
&= (N+\lambda_0)\mu^2
   -2\mu\bigl(N\bar y + \lambda_0\mu_0\bigr)
   +\text{const},
\end{aligned}

where $\displaystyle \bar y \;=\;\frac1N\sum_{n=1}^{N}y_n$.

#### 3. Insert back into $\log q^{*}(\mu)$  

$$
\boxed{%
\log q^{*}(\mu)
= -\frac{\bar\tau\,(N+\lambda_0)}{2}\,\mu^{2}
  \;+\;\bar\tau\,\mu\bigl(\lambda_0\mu_0 + N\bar y\bigr)
  \;+\;\text{const}}
$$

which is exactly equationÂ (7) of the exercise.

# 5.1â€¯h

Equationâ€¯(11) (from partâ€¯(g)) is  

$$
\log q^{*}(\tau)
= -\frac{\tau}{2}
   \Bigl[
     \sum_{n=1}^{N}(y_n-m)^{2}
     + \lambda_0(\mu_0-m)^{2}
     + (N+\lambda_0)s^{2}
   \Bigr]
   \;-\; b_0\tau
   +\Bigl(a_0+\frac{N+1}{2}-1\Bigr)\log\tau
   + \text{const}.
$$

---

#### Compare with the logâ€‘Gamma template  

For a Gamma distribution  
$$
\operatorname{Gam}(\tau\mid a,b)
  \;=\;
  \frac{b^{a}\,\tau^{a-1}}{\Gamma(a)}\,e^{-b\tau},
$$
we have  
$$
\log \operatorname{Gam}(\tau\mid a,b)
  \;=\;
  (a-1)\log\tau \;-\; b\tau \;+\; \text{const}.
$$

---

#### Match coefficients  

* **Coefficient of $\log\tau$**

  $$
  a-1
  \;=\;
  a_0 + \frac{N+1}{2} - 1
  \quad\Longrightarrow\quad
  \boxed{\,a = a_0 + \dfrac{N+1}{2}\,}.
  $$

* **Coefficient of $\tau$**

  $$
  b
  \;=\;
  b_0 + \tfrac12
      \Bigl[
        \sum_{n=1}^{N}(y_n-m)^{2}
        + \lambda_0(\mu_0-m)^{2}
        + (N+\lambda_0)s^{2}
      \Bigr],
  $$

  hence  

  $$
  \boxed{\,b = b_0
      + \dfrac12
        \Bigl(
          \sum_{n=1}^{N}(y_n-m)^{2}
          + \lambda_0(\mu_0-m)^{2}
          + (N+\lambda_0)s^{2}
        \Bigr)\,}.
  $$

---

Therefore  

$$
q^{*}(\tau) \;=\; \operatorname{Gam}\!\bigl(\tau \,\big|\, a,\; b\bigr),
$$

with parameters  

$$
a = a_0 + \frac{N+1}{2},
\qquad
b = b_0
    + \frac12\!
      \left(
        \sum_{n=1}^{N}(y_n-m)^{2}
        + \lambda_0(\mu_0-m)^{2}
        + (N+\lambda_0)s^{2}
      \right).
$$

#Â 5.1â€¯i

Start from the variational freeâ€‘energy 

$$
\mathcal F
= \mathbb E_{q(\theta)}\bigl[\log p(\mathbf y,\theta)\bigr]
  \;-\;
  \mathbb E_{q(\theta)}\bigl[\log q(\theta)\bigr],
  \qquad
  \theta=(\mu,\tau),\;
  q(\theta)=q(\mu)\,q(\tau).
$$

Collecting the $\mu$â€‘ and $\tau$â€‘dependent terms (and grouping constants) one obtains  

$$

\begin{aligned}
\mathcal F
&= -\frac{1}{2}\,\mathbb E_{\tau}[\tau]
     \!\left(
        2b_0
        + \mathbb E_{\mu}\!\Bigl[
            \sum_{n=1}^{N}(y_n-\mu)^2
            + \lambda_0(\mu-\mu_0)^2
          \Bigr]
      \right) \\[6pt]
&\quad
  +\left(a_0+\frac{N+1}{2}-1\right)\mathbb E_{\tau}[\log\tau]
  \;-\;\frac{N+1}{2}\log 2\pi
  \;-\;\log\Gamma(a_0) \\[6pt]
&\quad
  +\,a_0\log b_0
  \;+\;\tfrac12\log\lambda_0
  \;-\;\mathbb E_{\mu}[\log q(\mu)]
  \;-\;\mathbb E_{\tau}[\log q(\tau)]\;
\end{aligned}
$$

which is exactly equationÂ (14) in the exercise sheet.

Because in the normalâ€‘gamma prior the variance ofâ€¯$\mu$ after integrating outâ€¯$\tau$ is

$$
\operatorname{Var}[\mu]
\;=\; \mathbb E_{\tau}\!\bigl[\frac{1}{\lambda_0\tau}\bigr]
\;=\; \frac{b_0}{a_0\,\lambda_0},
$$

where we used $\mathbb E[\tau]=a_0/b_0$.
Initializing $s^2$ to this value makes the meanâ€‘field factor $q(\mu)$
consistent with the marginal prior and places the very first ELBO exactly at the
prior value.  Starting with $s^2=1/\lambda_0$ would instead correspond to
conditioning on a fixed $\tau=1$, which is incompatible with the prior and
forces an immediate jump in the first update step.
